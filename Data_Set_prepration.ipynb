I have used TwelveData API within date 2021-10-01 to 2021-10-20 in 5 minutes interval for BitCoin Prediction.

```python
import requests
import numpy as np
import pandas as pd
```

```python
api_key='61a8c16b48214b219374ac7cdedd419b'
symbol='BTC/USD'
interval='5min'
order='asc'
start_date1='2021-10-01 00:00:00'
end_date1='2021-10-20 00:00:00'

api_url=f"https://api.twelvedata.com/time_series?symbol={symbol}&start_date={start_date1}&end_date={end_date1}&interval={interval}&order={order}&apikey={api_key}"
```

```python
data=requests.get(api_url).json()
data
```
